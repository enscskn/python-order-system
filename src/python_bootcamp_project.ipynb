```python
import csv
import datetime
import qrcode
import random
import time

## pizza superclass
class Pizza:
    def get_description(self): 
        return self.__class__.__name__

    def get_cost(self): 
        return self.__class__.cost

## pizza subclass
class Klasik(Pizza):
    cost = 8.0

    def __init__(self):
        #Her alt sınıf için özellikler yazılacak.
        self.description = "Klasik Malzemeleri: Sucuk, Salam, Kaşar"
        print(self.description +"\n")

class Margarita(Pizza):
    cost = 5.0

    def __init__(self):
        self.description = "Margarita Malzemeleri: Domates, Fesleğen, Mozarella"
        print(self.description +"\n")

class TurkPizza(Pizza):
    cost = 10.0

    def __init__(self):
        self.description = "TürkPizza Malzemeleri: Kıyma, Soğan, Biber, Sarımsak, Kaşar"
        print(self.description +"\n")

class PlainPizza(Pizza):
    cost = 9.0


    def __init__(self):
        self.description = "PlainPizza Malzemeleri: Salam, Sucuk, Zeytin"
        print(self.description +"\n")


## decorator superclass
class Decorator(Pizza):
    def __init__(self, topping):
        self.component = topping

    def get_cost(self):
        return self.component.get_cost() + \
          Pizza.get_cost(self)

    def get_description(self):
        return self.component.get_description() + \
          ' + ' + Pizza.get_description(self)


## decorator subclass
class Zeytin(Decorator):
    cost = 0.5

    def __init__(self, topping):
        Decorator.__init__(self, topping)


class Mantar(Decorator):
    cost = 0.7

    def __init__(self, topping):
        Decorator.__init__(self, topping)


class Peynir(Decorator):
    cost = 1.0

    def __init__(self, topping):
        Decorator.__init__(self, topping)


class Et(Decorator):
    cost = 3.0

    def __init__(self, topping):
        Decorator.__init__(self, topping)


class Sogan(Decorator):
    cost = 0.2

    def __init__(self, topping):
        Decorator.__init__(self, topping)


class Misir(Decorator):
    cost = 0.4

    def __init__(self, topping):
        Decorator.__init__(self, topping)

## loading func 
def loading_screen():
    print("Sipariş hazırlanıyor...")
    for i in range(10):
        time.sleep(0.5) 
        print("[{:<10}] {:.0%}".format("="*i, (i+1)/10), end="\r", flush=True) 
    print("\nSiparişiniz Onaylandı.")

## main func for menu 
def main():
    with open("menu.txt") as cust_menu:
        for i in cust_menu:
            print(i, end="")

    class_dict = {1: Klasik, 
                  2: Margarita, 
                  3: TurkPizza, 
                  4: PlainPizza, 
                  5: Zeytin, 
                  6: Mantar, 
                  7: Peynir, 
                  8: Et, 
                  9: Sogan, 
                  10: Misir}
    
    code = input("Lütfen Pizzanızı Seçiniz: ")
    while code not in ["1", "2", "3", "4"]:
        code = input("\nYanlış tuşlama, tekrar deneyin!: ")

    order = class_dict[int(code)]()

    code = ""
    while code != "*":
        code = input('Ek malzeme için tuşlama yapınız (Ek malzeme istemiyorsanız "*" karakterini tuşlayabilirsiniz.): ')
        if code in ["5","6","7","8","9","10"]:
            order = class_dict[int(code)](order)
            print("\nMalzeme başarıyla eklendi!")
        elif code != "*":
            print("\nYanlış tuşlama, tekrar deneyin!: ")


    print("\n"+order.get_description().strip() +
          " = $" + str(order.get_cost()))

## drink recommendation
    pizza_choice = input("Hangi pizzayı seçtiğini söylersen sana bir içecek önerebiliriz (Klasik, Margarita, TürkPizza, PlainPizza): ")

    while pizza_choice not in ["Klasik", "Margarita", "TürkPizza", "PlainPizza"]:
        print("Hatalı giriş! Lütfen listeden bir pizza seçin.")
        pizza_choice = input("Hangi pizzayı seçtiğini söylersen sana bir içecek önerebiliriz (Klasik, Margarita, TürkPizza, PlainPizza): ")

    if pizza_choice == "Klasik":
        recommended_drink = "Kola"
        recommended_cost = "6"
    elif pizza_choice == "Margarita":
        recommended_drink = "Limonata"
        recommended_cost = "4"
    elif pizza_choice == "TürkPizza":
        recommended_drink = "Ayran"
        recommended_cost = "3"
    elif pizza_choice == "PlainPizza":
        recommended_drink = "IceTea"
        recommended_cost = "5"
    else:
        recommended_drink = None
        recommended_cost = None


    print("Önerilen içecek: " + recommended_drink)
    print()

    chooice = input("Siparişinize önerilen içeriği eklemek ister misiniz? (y/n)")
    while chooice not in ["y","n"]:
        print("Yanlış bir seçim yaptın!")
    if chooice in ["y"]:
        toplam_fiyat = float(order.get_cost()) + float(recommended_cost)

        print("Toplam ödemeniz gereken tutar: " + str(toplam_fiyat)+"$")
    else:
      recommended_drink = None
      recommended_cost = None
      print("Ödemen gereken tutar: "+ str(order.get_cost())+"$")



## login user information
    print("Lütfen Sipariş Bilgilerinizi Giriniz: \n")
    name = input("İsminiz: ")
    ID = input("TC Kimlik Numaranız: ")
    kk_no = input("Kredi Kartı Numaranızı Giriniz: ")
    kk_pw = input("Kredi Kartı Şifrenizi Giriniz: ")
    time_of_order = datetime.datetime.now()

    with open('Orders_Database.csv', 'a') as orders:
      orders = csv.writer(orders, delimiter=',')
      if recommended_drink != None:
        orders.writerow([name, ID, kk_no, kk_pw, order.get_description(), f"{order.get_cost()}$", recommended_drink, f"{recommended_cost}$", time_of_order])
      else:
        orders.writerow([name, ID, kk_no, kk_pw, order.get_description(), f"{order.get_cost()}$", time_of_order])


## security code generation
    random_number = random.randint(100000, 999999)
    print(random_number) #tester code
## qrcode generation
    qr = qrcode.QRCode(version=1, box_size=10, border=5)
    qr.add_data(str(random_number))
    qr.make(fit=True)
    qr.print_ascii()

## verification
    for i in range(3):
        user_input = input("Lütfen yukarıdaki QR kodu okut ve doğrulama kodunu girin: ")
        if user_input == str(random_number):
            print("Doğrulandı.")
            loading_screen()
            
            break
        else:
            if i == 2:
                print("Hesabınız bloke oldu.")
                break
            else:
                print(f"Yanlış kod girdiniz. Kalan deneme hakkınız: {2-i}")
 


main()
```
